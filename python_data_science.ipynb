#### Tabela Principal

##### Inicializando tabela principal

````python
# Criando a var df e lendo o excel na tabela Principal
df_principal = pd.read_excel("/content/planilha-imersao-python.xlsx", sheet_name="Principal")

# Visualizando as 5 primeiras linhas da tabela
print("----------Visualizando as 5 primeiras linhas da tabela----------")
head_visu = df_principal.head()
print(head_visu)
print("\n\n")

# Visualizando as 5 últimas linhas da tabela
print("----------Visualizando as 5 últimas linhas da tabela----------")
tail_visu = df_principal.tail()
print(tail_visu)
print("\n\n")

# Visualizando uma linha aleatória da tabela
print("----------Visualizando uma linha aleatória da tabela----------")
sample_visu = df_principal.sample()
print(sample_visu)
````

##### Configurando tabela principal
- O intuito desse trecho é criar as colunas igual as do Excel. Para isso utilizaremos as funções do python para nos auxiliar



- Definindo quais colunas serão copiadas de **df_principal**
- Renomeando-as para melhor compreensão

````python
# Pegando colunas específicas
df_principal = df_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()

# Renomeando colunas
df_principal = df_principal.rename(columns={"Último (R$)": "valor_final", "Var. Dia (%)": "var_dia_pct"}).copy()

# Visualizando as 5 primeiras linhas da nova tabela
print("----------Visualizando as 5 primeiras linhas da tabela----------")
head_visu = df_principal.head()
print(head_visu)
print("\n\n")
````



---


- Criando a coluna **"var_pct"** que terá a função de conter o valor em porcentagem da variação da empresa no dia da base de dados

````python
df_principal["var_pct"] = df_principal["var_dia_pct"]/100
````



---


- Criando a coluna **"valor_inicial"** que terá a função de conter o valor da cotação daquela empresa no começo do dia.
- Utilizaremos a coluna criada na etapa anterior e utilizaremos a função nativa do Python **round()** para arredondar em 2 casa decimais o valor incial




````python
df_principal["valor_inicial"] = round(df_principal["valor_final"]/(df_principal["var_pct"] + 1), 2)
````



---


- Criando a coluna **"qndt_teorica"** que terá a função de conter o valor em porcentagem da variação da empresa no dia da base de dados
- Para isso utilizaremos a função merge, que uni duas planilhas a parti de colunas em comum


```
df_principal = df_principal.merge(planilha_a_ser_copiada, left_on="coluna_de_df_principal",
                                    right_on="coluna_da_planila_copiada", how="tipo_de_join")

# how="left" indica que estamos realizando uma junção do tipo "left join", ou seja, todas as linhas do DataFrame à esquerda (df_principal)
serão mantidas, e as colunas do DataFrame à direita (df_total_acoes) serão adicionadas quando houver correspondência.
```

- Após esse merge precisaremos excluir a coluna **"Código"** que veio da planilha **df_total_acoes**. Para isso utilizaremos a função **drop()** do Python

- Por fim, renomearemos a coluna **qtnd_teorica** para que ela fique de acordo com as outras

````python
# Criando coluna da quantidade de ações, para isso usamos a função merge()
df_principal = df_principal.merge(df_total_acoes, left_on="Ativo", right_on="Código", how="left")

# Removendo a coluna duplicada
df_principal = df_principal.drop("Código", axis=1)

# Renomeando coluna nova criada
df_principal = df_principal.rename(columns={"Qtde. Teórica": "qntd_teorica"}).copy()
````



---


- Criando a coluna **"var_reais"** que terá a função de conter a variação em reais das ações daquela empresa

- Também atualizaremos o formato da planilha para que os valores que sejam float tenha 2 casa decimais

- Por fim iremos definir a tipo da coluna **"qntd_teorica"** para inteiro, pois se é quantidade é algo sem casa decimais

````python
# Criando coluna de variação em reais
df_principal["var_reais"] = (df_principal["valor_final"] - df_principal["valor_inicial"])*df_principal["qntd_teorica"]

# Atualizando a formato da planilha para float
pd.options.display.float_format = '{:.2f}'.format

# Tranformando a coluna qntd_teorica de float para inteiro
df_principal["qntd_teorica"] = df_principal["qntd_teorica"].astype(int)
````



---


- Mostrando resultado atual da planilha

````python
df_principal.head()
````



---


- Criando a coluna **"resultado"** que conterá a informação se as ações daquela empresa subiram, desceram ou ficaram estáveis naquele dia

- Utilizaremos a função **apply()** que irá percorrer cada elemento da coluna **"var_reais"** e o lamda para especificar o que será feito com cada elemento

````python
df_principal["resultado"] = df_principal["var_reais"].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estvável"))
````



---


- Criando a coluna **"nome_empresa"**, **"segmento_empresa"**, **"idade_empresa"** que conterá o nome da empresa escrito por extenso, o segmento dela e a idade da empresa

- Utilizaremos novamente o merge, mas dessa vez faremos que apenas a coluna de interesse seja pega

````python
# Criando a coluna nome da empresa
df_principal["nome_empresa"] = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how="left")["Nome"]

# Criando a coluna segmento da empresa
df_principal["segmento_empresa"] = df_principal.merge(df_chatgpt, left_on="nome_empresa", right_on="Nome da empresa", how="left")["Segmento"]

# Criando a coluna idade da empresa
df_principal["idade_empresa"] = df_principal.merge(df_chatgpt, left_on="nome_empresa", right_on="Nome da empresa", how="left")["Idade da Empresa (em 2024)"]
````



---


- Criando a coluna **"cat_empresa"** que conterá a categoria da empresa em relação a idade dela

````python
df_principal["cat_empresa"] = df_principal["idade_empresa"].apply(lambda x: "Mais de 100" if x > 100 else ("Menos de 50" if x < 50 else "Entre 50 e 100"))

df_principal.head()
````

### Análises em cima dos dados



---


- Procurando o valor máximo, o valor mínimo e a média da planilha principal e calculando a média as variações que desceram e que subiram

````python
maior_variacao = df_principal['var_reais'].max()
menor_variacao = df_principal['var_reais'].min()
media_variacao = df_principal['var_reais'].mean()

media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_reais'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_reais'].mean()

print(f"Maior variação:\t R${maior_variacao: ,.2f}")
print(f"Menor variação:\t R${menor_variacao: ,.2f}")
print(f"Média das variações:\t R${media_variacao: ,.2f}")
print(f"Média das variações que subiu:\t R${media_subiu: ,.2f}")
print(f"Média das variações que desceu:\t R${media_desceu: ,.2f}")
````



---


- Criando um novo df com as empresas que as ações subiram e dividindo por segmento

````python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu.head(5)
````

````python
df_analise_segmento = df_principal_subiu.groupby("segmento_empresa")["var_reais"].sum().reset_index()
df_analise_segmento.head(3)
````

````python
df_analise_saldo = df_principal.groupby("resultado")["var_reais"].sum().reset_index()
df_analise_saldo.head(3)
````

````python
df_analise_idade = df_principal.groupby("cat_empresa")["var_reais"].sum().reset_index()
df_analise_idade["qntd_empresas"] = df_principal.groupby("cat_empresa")["nome_empresa"].count().values
df_analise_idade.head(3)
````

### Criando gráficos

#### Gráfico de barras

````python
# Criar o gráfico de barras com personalizações
figura = px.bar(
    df_analise_saldo,
    x="resultado",
    y="var_reais",
    text="var_reais",
    title="Saldo das Ações",
    color="resultado",
    color_discrete_sequence=["red", "white", "green"]
)

# Melhorar a apresentação do gráfico
figura.update_traces(
    texttemplate='%{text:.2f}',  # Formato dos números nos rótulos das barras
    textposition='outside'       # Posição dos rótulos fora das barras
)

figura.update_layout(
    title={
        'text': "Saldo das Ações",
        'y':0.9,  # Posição do título no eixo y
        'x':0.5,  # Posição do título no eixo x
        'xanchor': 'center',  # Ancoragem do título
        'yanchor': 'top'      # Ancoragem do título
    },
    xaxis_title="Resultado",
    yaxis_title="Variação em Reais",
    yaxis_tickprefix="R$ ",
    yaxis_tickformat=",.0f",  # Formato dos números no eixo y
    uniformtext_minsize=8,    # Tamanho mínimo do texto
    uniformtext_mode='hide',  # Modo de exibição do texto
    plot_bgcolor='rgba(0,0,0,0)',  # Fundo do gráfico transparente
    bargap=0.2  # Espaço entre as barras
)

# Exibir o gráfico
figura.show()
````
